```python
k1 = GPflow.kernels.RBF(1)
k2 = GPflow.kernels.PeriodicKernel(1,period=15.0)
X=np.arange(0,df_tmp['xx'].size,dtype=float).reshape(df_tmp['xx'].size,1)[0:100]
Y=df_tmp['xx'].values.reshape(df_tmp['xx'].size,1)[0:100]
plt.plot(X, Y, 'kx', mew=2)
```

```python

k = GPflow.kernels.RBF(1, lengthscales=0.3)
m = GPflow.gpr.GPR(X, Y, kern=k1*k2)
m.likelihood.variance = 0.01
```

```python
def plot(m):
    xx = np.arange(0, 168,1,dtype=float)[:,None]
    mean, var = m.predict_y(xx)
    plt.figure(figsize=(12, 6))
    plt.plot(X, Y, 'kx', mew=2)
    plt.plot(xx, mean, 'b', lw=2)
    plt.fill_between(xx[:,0], mean[:,0] - 2*np.sqrt(var[:,0]), mean[:,0] + 2*np.sqrt(var[:,0]), color='blue', alpha=0.2)
    plt.xlim(0, 168)
plot(m)
```

```python
m.optimize(maxiter=10000)
plot(m)
m
```
